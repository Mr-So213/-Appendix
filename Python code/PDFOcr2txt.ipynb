```python
from PIL import Image
import sys
import pyocr
import pyocr.builders
import pdf2image
import glob
from tqdm import tqdm
from zipfile import ZipFile
```

```python
tools = pyocr.get_available_tools()
if len(tools) == 0:
    print("No OCR tool found")
    sys.exit(1)

# The tools are returned in the recommended order of usage
tool = tools[0]
print("Will use tool '%s'" % (tool.get_name()))
# Ex: Will use tool 'libtesseract'
```

```python
def pdf2text(path):
  images = pdf2image.convert_from_path(path, dpi=200, fmt='jpg')
  lang = 'eng'
  text_lst = []
  #img object to txt file and output it
  for image in images:
    text = tool.image_to_string(
        image,
        lang=lang,
        builder=pyocr.builders.TextBuilder())
    text_lst.append(text)
  
  texts = "\n".join(text_lst)
  
  return texts
```

```python
def texts2txt(path, texts):
  path_new = path.replace(".pdf", ".txt")
  f = open(path_new, "w")
  f.write(texts)
  f.close()
```

```python
def pdf2txt(path):
  texts = pdf2text(path)
  texts2txt(path, texts)
```

```python
def getPDFpaths(folderpath):
  path_lst = glob.glob(folderpath + "/*.pdf")
  return path_lst
```

```python
file_name = "/content/before2011.zip"

with ZipFile(file_name, 'r') as zip:
  zip.extractall()
  print('Done')
```

```python
def main():
  folderpath = input("please copy and paste the folder's path to here: ")
  path_lst = getPDFpaths(folderpath)
  for path in tqdm(path_lst):
    pdf2txt(path)
```

```python
main()
```
