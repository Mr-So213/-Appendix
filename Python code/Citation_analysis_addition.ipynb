```python
import pandas as pd
from collections import Counter
import re
```

```python
def citation_count(path):
  df = pd.read_excel(path)
  communities = sorted(list(set(df["Community"])))
  res_lst = []
  for community in communities:
    df_sub = df[df["Community"] == community]
    if len(df_sub) > 1:
      refer_lst = []
      for references in df_sub["Cited References"]:
        lst = references.split("; ")
        refer_lst = refer_lst + lst
      
      counter = Counter(refer_lst)
      res = sorted(counter.items(), key=lambda i: i[1], reverse=True)

      res_lst.append(res)
    
    elif len(df_sub) == 1:
      refer_lst = []
      for references in df_sub["Cited References"]:
        lst = references.split("; ")
        refer_lst = refer_lst + lst
      
      counter = Counter(refer_lst)
      res = sorted(counter.items(), key=lambda i: i[1], reverse=True)
      res_new = [(i[0], "only1article") for i in res]

      res_lst.append(res_new)
  
  df_output = pd.DataFrame()
  #df_output["INDEX"] = list(range(detect_len(res_lst)))
  for num, community in enumerate(communities):

    df_output["Reference_community" + str(community)] = pd.Series([i[0] for i in res_lst[num]])
    df_output["Count_community" + str(community)] =  pd.Series([i[1] for i in res_lst[num]])

  df_output.to_excel(path.replace(".xlsx", "_communities.xlsx"), index=False)
  return df_output
```
